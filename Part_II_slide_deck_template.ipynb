## Dataset Overview

> The Ford GoBike data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area.This data collected from Feb 2019.

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
bike_df = pd.read_csv('ford_gobike.csv')
````

##  The Gender and age group that are more likely to use GoBike 

As we can see that the most of GoBike riders are males and their age group are between 20 - 40 years old

````python
plt.figure(figsize = [16,6])

plt.subplot(1,2,1)
color_base = sb.color_palette()[9]
order_gender = bike_df.member_gender.value_counts().index # order based on value 
sb.countplot(data = bike_df, x = 'member_gender', color = color_base, order = order_gender);  
plt.title('The gender of GoBike riders ', fontsize = 14);
plt.xlabel('member gender', fontsize = 12);
plt.ylabel('Number of rides', fontsize = 12);

plt.subplot(1,2,2)
bins = np.arange(0,bike_df.member_age.max()+5 , 5)
plt.hist(data = bike_df , x = 'member_age' , bins = bins);
plt.title('The members ages of GoBike riders ', fontsize = 14);
plt.xlabel('The members Age (years)', fontsize = 12);
plt.ylabel('Number of rides', fontsize = 12);
plt.xlim((0,80));
````

## Duartion in minute for each day of a week to each user type

the chart showed that Saturday and Sunday having the most duration time in minute.The Subscribers were having a less duration time compared to Customers.Which means that Customers are more likely to spends more time in a trip.

````python
plt.figure(figsize = [16,8])
weekdays = ['Sunday', 'Monday', 'Tuesday', 'Wednesday','Thursday','Friday', 'Saturday']
sb.boxplot(data = bike_df, x = 'start_day', y = 'duration_min', hue = 'user_type', order = weekdays)
plt.ylim((0,40));
plt.legend(loc = 6 , bbox_to_anchor = (1.0,0.5));
plt.title('Duartion in minute for each day of a week to each user type', fontsize =16);
plt.xlabel('Start rental day', fontsize = 12);
plt.ylabel('duration in (mintue)', fontsize = 12);
````

## The Start Rentals hour and day for Customers

the chart show that the Start hour of the weekdays different from weekends days with a high demands in the afternoon and above while in the weekdays the high demands are at 8am and 5pm. 

````python
bike_df_sub = bike_df[bike_df['user_type'] == 'Customer']
bins = np.arange(0, 24 , 1)
g = sb.FacetGrid( data = bike_df_sub , col = 'start_day' ,col_wrap = 4, size = 3,sharey = True)
g = g.map(plt.hist , 'start_hour',edgecolor='k', bins = bins)
plt.xlim((0,24)); # axis limit
plt.suptitle('The Start Rentals hour and day for Customers ', y =1.05, fontsize = 14);
````

## The Start Rentals hour and day for Subscriber 

As we can see from the chart which is very interesting is that the weekend days is the lowest Number of rentals for the Subscribers which could means that the most of the Subscribers used the GoBike system for their work and also with the same pattern in the rental start hour in the weekdays.

````python
bike_df_sub = bike_df[bike_df['user_type'] == 'Subscriber']
bins = np.arange(0, 24 , 1)
g = sb.FacetGrid( data = bike_df_sub , col = 'start_day', col_wrap = 4, size = 3,sharey = True)
g = g.map(plt.hist , 'start_hour',edgecolor='k', bins = bins)
plt.xlim((0,24));
plt.suptitle('The Start Rentals hour and day for Subscriber ', y =1.05, fontsize = 14);
````
