## Cleaning 

There are some Qualtiy issues needs to be cleaned before we go further with the analysis:

- There are some missing values in 6 columns within the data set 
- Ther start and end time coulmns are not 'timestamp' data type
- The member birth year is float
- The columns : start_station_id , end_station_id and bike_id need to in object format.
- Adding Member Age column From the Birth year
- Extract The Days and Hours of the start and end time columns to help us with further analysis
- Extract a column for a duration in minute to faciliate further analysis

### Define :  Drop the missing values that represents  a few rows that won’t affect our analysis.
### code:

```python
(bike_df.isnull().sum()).sum() / bike_df.shape[0]  # due to the low missing proportion, the missing values will be dropped   
```

```python
bike_df.dropna(inplace = True) # Drop the missing values 
```

### test:

```python
bike_df.isnull().sum() # The missing values dropped successfully 
```

### Define: Change the start and end time coulmns to a timestamp datatype

### code:

```python
# convert the dataype to a datetime
bike_df.start_time = pd.to_datetime(bike_df.start_time) 
bike_df.end_time = pd.to_datetime(bike_df.end_time)
```

### test:

```python
bike_df.info()
```

```python
bike_df.start_time.min() # the first day
```

```python
bike_df.start_time.max() # the last day in the data set
```

### Define: convert The member birth column to an integer type

### code:

```python
bike_df.member_birth_year = bike_df.member_birth_year.astype(int) # convert datatype to integer 
```

### test:

```python
bike_df.member_birth_year.dtypes # the data type converted successfully
```

### Define: change the columns (start_station_id , end_station_id and bike_id ) to an object format.

### code:

```python
bike_df.start_station_id = bike_df.start_station_id.astype(str)
bike_df.end_station_id = bike_df.end_station_id.astype(str)
bike_df.bike_id = bike_df.bike_id.astype(str)
```

### test:

```python
bike_df.info() # the data types converted successfully
```

### Define : Adding Member Age column From the Birth year
### code :

```python
bike_df['member_age'] = 2019 - bike_df['member_birth_year'] # add a new column for member age for easier exploration
```

### test:

```python
bike_df.head()
```

### Define : Extract The Days and Hours of the start and end time columns to help us with further analysis
### code:

```python
bike_df['start_day'] = bike_df['start_time'].dt.strftime('%A') # extract The day 
bike_df['start_hour'] = bike_df['start_time'].dt.strftime('%H').astype(int)

bike_df['end_day'] = bike_df['end_time'].dt.strftime('%A')
bike_df['end_hour'] = bike_df['end_time'].dt.strftime('%H').astype(int)
```

### test:

```python
bike_df.sample(5) # test 
```

### Define : Extract a column for a duration in minute to faciliate further analysis
### code:

```python
bike_df['duration_min'] = bike_df['duration_sec'] / 60 # create a new column for thb duration minute
```

### test :

```python
bike_df.head() # test
```

```python
bike_df.to_csv('ford_gobike.csv', index = False)
```

## Univariate Exploration

### Which gender are more likely to use GoBike ?

```python
# creating a count plot for the number of rides for each genders 
plt.figure(figsize = [10,6]);
color_base = sb.color_palette()[9]
order_gender = bike_df.member_gender.value_counts().index # order based on value 
sb.countplot(data = bike_df, x = 'member_gender', color = color_base, order = order_gender);  
plt.title('Which gender are more likely to use GoBike ?', fontsize = 14);
plt.xlabel('member gender', fontsize = 12);
plt.ylabel('Number of rides', fontsize = 12);
```

As we can see from the graph that the number of males are quite large compared to females. so, we can say that males are more likely to use GoBike.

### Subscriber user Vs. Customer user 

```python
# creating a pie chart to show the proportion of each user type 
sort_counts = bike_df.user_type.value_counts() 
plt.figure(figsize = [12,6]);
plt.pie(sort_counts, autopct = '%1.1f%%' ,labels = sort_counts.index ,  startangle = 90 , counterclock = False );
plt.axis('square');
plt.title('Subscriber user Vs. Customer user', fontsize = 14);
```

As we can see from the graph that Subscriber user has a high proportion compared to Customer user type with a 90.5% 

### GoBike Rentals Duration in minute  

```python
# creating a histogram to show the rentals duration for bike rides in minute
bins = np.arange(0 , 40 +1 , 1)
plt.figure(figsize = [12,6])
plt.hist(data = bike_df , x = 'duration_min', bins = bins);
plt.title('GoBike Rental Duration in minute', fontsize = 14);
plt.xlabel('duration (min)', fontsize = 12);
plt.ylabel('Number of rides', fontsize = 12);
```

As it shown from the graph above that rental duration is a right skewed and most rides fall in between 4 - 12 minute . While very few rides reach half an hour.

### The Members Ages of GoBike riders 

```python
# creating a histogram of members ages of riders
bins = np.arange(0,bike_df.member_age.max()+5 , 5)
plt.figure(figsize = [12,6])
plt.hist(data = bike_df , x = 'member_age' , bins = bins);
plt.title('The members ages of GoBike riders ', fontsize = 14);
plt.xlabel('The members Age (years)', fontsize = 12);
plt.ylabel('Number of rides', fontsize = 12);
plt.xlim((0,80));
```

As it shown from the graph above the majority of the members ages are between 20 - 40 years. 

### The Number of users enrolled The Share Bike for all trip 

```python
# creating a count plot for the numbers of enrolled in share bake 
plt.figure(figsize = [8,6]);
color_base = sb.color_palette()[9]
order_gender = bike_df.bike_share_for_all_trip.value_counts().index
sb.countplot(data = bike_df, x = 'bike_share_for_all_trip', color = color_base, order = order_gender);
plt.title('The number of users enrolled The Share Bike for all trip', fontsize = 14);
plt.xlabel('bike share for all trip', fontsize = 12);
plt.ylabel('Number of rides', fontsize = 12);
```

As it shown from the above chart the most of the users wasn’t enrolled the share bike for all trip.

### What day of the week has the heighest number of rentals ?

As it shown from the above chart that Thursday is having the highest number of rentals, while the weekends (sunday and saturday) are the least.

### Which hour was more likely to a user to start up a trip ?

### The Start Rentals hour and day for Customers 

```python
# creating a FacitGrid for the start day and hour for the customers
bike_df_sub = bike_df[bike_df['user_type'] == 'Customer']
bins = np.arange(0, 24 , 1)
g = sb.FacetGrid( data = bike_df_sub , col = 'start_day' ,col_wrap = 3, size = 4,sharey = True)
g = g.map(plt.hist , 'start_hour',edgecolor='k', bins = bins)
plt.xlim((0,24)); # axis limit
plt.suptitle('The Start Rentals hour and day for Customers ', y =1.05, fontsize = 14);
```

As we can see from the above chart is that the Start hour of the weekdays different from weekends days with a high demands in the afternoon and above while in the weekdays the high demands are at 8am and 5pm. 

### The Start Rentals hour and day for Subscriber 

```python
# creating a FacitGrid for the start day and hour for the Subscriber
bike_df_sub = bike_df[bike_df['user_type'] == 'Subscriber']
bins = np.arange(0, 24 , 1)
g = sb.FacetGrid( data = bike_df_sub , col = 'start_day', col_wrap = 3, size = 4,sharey = True)
g = g.map(plt.hist , 'start_hour',edgecolor='k', bins = bins)
plt.xlim((0,24));
plt.suptitle('The Start Rentals hour and day for Subscriber ', y =1.05, fontsize = 14);
```

As we can see from the above chart which is very interesting is that the weekend days is the lowest Number of rentals for the Subscribers which could means that the most of the Subscribers used the GoBike system for their work and also with the same pattern in the rental start hour in the weekdays.
